```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
import numpy as np
from adjustText import adjust_text
```

```python
#input data

input = "{檔名.csv}"

df = pd.read_csv(input, sep=",")
df = df.apply(pd.to_numeric, errors="coerce")
df = df.dropna()

time = df[df.columns[0]]
freq = df[df.columns[1]]
```

```python
#fitting
mask = (time >={資料從哪邊開始(s)}) & (time <={資料從哪邊結束(s)})

time_fit = time[mask]
freq_fit = freq[mask]

fit = np.polyfit(time_fit, freq_fit, 1)
```

```python
plt.figure(figsize=(9,6))
#plt.title("Freq vs Time", fontsize=16)

plt.plot(time_fit, fit[0] * time_fit + fit[1], color='red', lw=1, label='fitting line')


plt.scatter(time, freq, color='black', lw=1, marker='.')
plt.xlim({資料從哪邊開始(s)}, {資料從哪邊結束(s)})
plt.ylim({資料從哪邊開始(Hz)}, {資料從哪邊開始(Hz)})
plt.xlabel("Time(s)", fontsize=14)
plt.ylabel("Frequency(Hz)", fontsize=14)
plt.legend()
plt.show()

print(f"slope:{fit[0]}")
```
