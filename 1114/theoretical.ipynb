```python
import numpy as np
import matplotlib.pyplot as plt
```

# 模擬

##### 40拍

```python
f0 = 1000        # 聲源發射頻率 (Hz)
R = 0.3          # 圓周半徑 (m)
omega = 2*np.pi*0.67  # 角速度 (rad/s) → 0.67 圈/秒
c = 340          # 聲速 (m/s)

# 觀察者位置（固定在圓外）
D = 0.98
observer = np.array([D, 0.0])

# 時間軸
t = np.linspace(0, 20, 2000)
```

```python
xs = R * np.cos(omega * t)
ys = R * np.sin(omega * t)
source_pos = np.vstack((xs, ys)).T

# 聲源切向速度 v = Rω
v = R * omega
vxs = -v * np.sin(omega * t)
vys =  v * np.cos(omega * t)
source_vel = np.vstack((vxs, vys)).T

# 從聲源指向觀察者的向量
vec = observer - source_pos
dist = np.linalg.norm(vec, axis=1)
unit_vec = vec / dist[:, None]

# 徑向速度 vr（正值 = 靠近觀察者）
vr = np.sum(source_vel * unit_vec, axis=1)
```

```python
f = f0 * c / (c - vr)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(t, f)
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.title("Doppler Frequency vs Time (Circular Motion)")
#plt.grid(True)
plt.show()
```
