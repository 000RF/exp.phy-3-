```python
import numpy as np
import matplotlib.pyplot as plt

# =================  =================
v_sound = 340.0        #  (m/s)
f_source_list = [1000, 1500, 2000]      #  (Hz)
distance = 2.0         #  (m)
g = 9.8                #  (m/s^2)

# PS
# 
# ...

# =================  =================

#
# h = 1/2 * g * t^2  => t = sqrt(2h/g)
total_time = np.sqrt(2 * distance / g)
 
t = np.linspace(0, total_time, 1000)

#  (v = gt)
v_receiver = g * t

# =================  =================
# 
for f_s in f_source_list:
    f_obs = f_s * ((v_sound + v_receiver) / v_sound)
    plt.figure(figsize=(10, 6))
    plt.plot(t, f_obs, label=f'Source:{f_s}Hz', lw=1, color='darkmagenta')
    plt.title('Linear Doppler Effect Simulation', fontsize=14)
    plt.xlabel('Time (s)', fontsize=12)
    plt.ylabel('Observed Frequency (Hz)', fontsize=12)
    plt.grid(True, linestyle='-', alpha=0.2)
    plt.legend()
    plt.show()
```
